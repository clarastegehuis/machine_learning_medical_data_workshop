# 🎨 Welkom bij de Convolutie Workshop!

---

## 👋 Hallo!

Welkom bij de workshop **Convolutie: Van Snapchatfilters tot beeldbewerking**.

In deze workshop leer je hoe de filters van Snapchat, Instagram en TikTok werken — en maak je ze zelf!

---

## 🚀 Hoe begin je?

### Stap 1: Open de workshop-notebook

Kijk naar de **linkerzijbalk** (de bestandsverkenner links in beeld). Klik daar op:

> 📄 **`workshop_convolutie_geen_ai_jupyterlite.ipynb`**

### Stap 2: Zie je de zijbalk niet?

Klik op het **mapje-icoon** linksboven (of druk op **Ctrl+Shift+F**) om de bestandsverkenner te openen.

### Stap 3: Aan de slag!

Zodra de notebook open is, begin je bij de eerste cel en volg je de instructies.

---

## 💡 Tips

- **Cel uitvoeren:** Klik op de cel en druk op **Shift + Enter**
- **Begin altijd bovenaan** en werk de cellen van boven naar beneden af
- Je kunt **je eigen foto uploaden** en er filters op toepassen!
